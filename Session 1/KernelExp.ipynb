```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
a= np.linspace(-3,3,100)
b= np.linspace(-100,100,100)

mean = 0.0
sigma = 1.0

# Don't worry about trying to learn or remember this formula.  I always have to refer to textbooks or check online for the exact formula.
z = (np.exp(np.negative(np.power(a - mean, 2.0) /
                   (2.0 * np.power(sigma, 2.0)))) *
     (1.0 / (sigma * np.sqrt(2.0 * 3.1415))))

mean = 0.0
sigma = 20.0
z1 = (np.exp(np.negative(np.power(b - mean, 2.0) /
                   (2.0 * np.power(sigma, 2.0)))) *
     (1.0 / (sigma * np.sqrt(2.0 * 3.1415))))
```

```python
plt.plot(z)
plt.show()
plt.plot(z1)
```
